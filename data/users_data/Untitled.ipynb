```python
import os
import pandas as pd
import sqlite3
```

```python
links = pd.read_csv('links.csv')
movies = pd.read_csv('movies.csv')
ratings = pd.read_csv('ratings.csv')
tags = pd.read_csv('tags.csv')
```

```python
connection = sqlite3.connect('movie_data.db')
```

```python
links.to_sql('links', con = connection, index = False)
movies.to_sql('movies', con = connection, index = False)
ratings.to_sql('ratings', con = connection, index = False)
tags.to_sql('tags', con = connection, index = False)
```

```python
connection.close()
```

```python
cursor = connection.cursor()
```

```python
query = "SELECT userId, movieId, rating FROM ratings"
res = cursor.execute(query)
ratings = res.fetchall()
```
